```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('BostonHousing.csv')
df
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df['rm'] = df['rm'].fillna(df['rm'].median())
df.isna().sum()
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(df.corr(), annot = True, cmap='coolwarm')
```

```python
df.columns
```

```python
df.drop(columns ='rad', inplace=True)
```

```python
df.columns
```

```python

df = df.rename(columns={'medv':'price' })
df
```

```python
X = df.drop(columns = 'price')
Y = df[['price']]
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state= 42)
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor(gamma=1,
    learning_rate=0.05,
    n_estimators=300,
    reg_lambda=20,
    reg_alpha=5,
    max_depth=3,
    random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import ShuffleSplit, cross_validate
import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')


shuffle_split = ShuffleSplit(n_splits=100, test_size=0.2, random_state=42)

cross_validate = cross_validate(model, X,Y, cv=shuffle_split, scoring='r2', return_train_score=True)

train_score = cross_validate['train_score']
test_score = cross_validate['test_score']

print(np.round(train_score.mean(),2), np.round(test_score.mean(),2))
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

train_score1 = r2_score(y_train, y_pred_train)
test_score2 = r2_score(y_test, y_pred_test)

train_score1_mae = mean_absolute_error(y_train, y_pred_train)
test_score2_mae = mean_absolute_error(y_test, y_pred_test)

train_score1_mse = mean_squared_error(y_train, y_pred_train)
test_score2_mse = mean_squared_error(y_test, y_pred_test)

print("r2 score is ", train_score1, "r2 score is ", test_score2)
print("mae score is ", train_score1_mae, "mae score is ", test_score2_mae)
print("mse score is ", train_score1_mse, "mse score is ", test_score2_mse)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(y_train, y_pred_train, alpha=0.6, color="blue")
plt.plot([y_train.min(), y_train.max()],
         [y_train.min(), y_train.max()],
         color="red", linestyle="--")  # 45° line

plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices (Train)")
plt.show()
```
